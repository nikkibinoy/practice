finding outliers and treating it

```python
import matplotlib.pyplot as plt
```

```python
df.boxplot()
plt.xticks(rotation=45)
plt.show()
```

```python
df.columns
```

```python
columns_to_fill = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History']
for column in columns_to_fill:
  plt.figure()
  df.boxplot([column])
  plt.title(column)
  plt.show()
```

```python
q1=np.percentile(df['ApplicantIncome'],25,interpolation='midpoint')
q2=np.percentile(df['ApplicantIncome'],50,interpolation='midpoint')
q3=np.percentile(df['ApplicantIncome'],75,interpolation='midpoint')
print(q1,q2,q3)
```

```python
iqr=q3-q1
print(iqr)
```

```python
min=q1-1.5*iqr
max=q3+1.5*iqr
print(min,max)
```

```python
outliers=[]
for x in df['ApplicantIncome']:
  if((x>max) or (x<min)):
    outliers.append(x)
print(outliers)
```

```python
df['ApplicantIncome']=df['ApplicantIncome'].clip(min,max)
```

```python
plt.boxplot(df['ApplicantIncome'])
plt.xlabel('ApplicantIncome')
plt.show()
```

```python
Q1=np.percentile(df['CoapplicantIncome'],25,interpolation='midpoint')
Q2=np.percentile(df['CoapplicantIncome'],50,interpolation='midpoint')
Q3=np.percentile(df['CoapplicantIncome'],75,interpolation='midpoint')
print(Q1,Q2,Q3)
```

```python
IQR=Q1-Q3
print(IQR)
```

```python
MIN=Q1-1.5*IQR
MAX=Q3+1.5*IQR
print(MIN,MAX)
```

```python
OUTLIERS=[]
for x in df['CoapplicantIncome']:
  if((x>MAX) or (x<MIN)):
    OUTLIERS.append(x)
print(OUTLIERS)
```

```python
df['CoapplicantIncome']=df['CoapplicantIncome'].clip(MIN,MAX)
```

```python
plt.boxplot(df['CoapplicantIncome'])
plt.xlabel('CoapplicantIncome')
plt.show()
```
